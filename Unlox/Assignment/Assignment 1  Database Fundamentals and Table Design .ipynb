## Part A: Theory

### 1. What is a database in MySQL? Why is it important to organize data into databases?

In MySQL, a **database** (also known as a schema) is a structured collection of tables, views, stored procedures, functions, and other objects that are logically grouped together. It acts as a container for related data.

**Importance of organizing data into databases:**
*   **Organization:** Provides a clear and logical way to group related data, making it easier to manage and understand.
*   **Security:** Allows for different access permissions to be set at the database level, enhancing data security.
*   **Data Integrity:** Helps maintain data consistency and integrity by defining relationships and constraints within a single database.
*   **Scalability:** Facilitates scaling by allowing different applications to use separate databases, or by distributing databases across multiple servers.
*   **Backup and Recovery:** Simplifies backup and recovery operations as entire databases can be backed up and restored independently.

### 2. Explain the difference between INT and BIGINT. When would you prefer one over the other?

Both `INT` and `BIGINT` are integer data types used to store whole numbers in MySQL, but they differ in the range of values they can store and the amount of storage they consume.

*   **INT (Integer):** Stores whole numbers. Its range is typically from -2,147,483,648 to 2,147,483,647 for signed values, and 0 to 4,294,967,295 for unsigned values. It usually occupies 4 bytes of storage.
*   **BIGINT (Big Integer):** Stores much larger whole numbers. Its range is typically from -9,223,372,036,854,775,808 to 9,223,372,036,854,775,807 for signed values, and 0 to 18,446,744,073,709,551,615 for unsigned values. It usually occupies 8 bytes of storage.

**When to prefer one over the other:**
*   **Prefer `INT`:** When the numbers you need to store are within its range. `INT` uses less storage and can be slightly more efficient for calculations.
*   **Prefer `BIGINT`:** When the numbers you anticipate storing might exceed the range of `INT`. This is common for primary keys in very large tables (e.g., auto-incrementing IDs that could grow beyond 2 billion), or for storing counts or monetary values that can reach very large figures.

### 3. What is a PRIMARY KEY? Mention two rules that a primary key must follow.

A **PRIMARY KEY** is a special column or a set of columns in a database table that uniquely identifies each record (row) in that table. It serves as a unique identifier for each instance of an entity in the table.

**Two rules a primary key must follow:**
1.  **Uniqueness:** Each value in the primary key column(s) must be unique across all rows in the table. No two rows can have the same primary key value.
2.  **Non-Nullability:** A primary key column (or any column part of a composite primary key) cannot contain `NULL` values. Every record must have a valid and non-null primary key value.

### 4. Differentiate between CHAR and VARCHAR with one practical example.

Both `CHAR` and `VARCHAR` are used to store string data, but they handle storage and space differently.

*   **CHAR (Fixed-Length String):**
    *   Stores a fixed-length string. If the stored string is shorter than the defined length, it is padded with spaces to fill the entire length.
    *   Uses a fixed amount of storage space (equal to its defined length) regardless of the actual string length.
    *   Retrieval can be slightly faster for fixed-length data.
    *   **Practical Example:** Storing two-letter state abbreviations (e.g., 'CA', 'NY'). If you define `CHAR(2)`, it will always store exactly 2 characters.

*   **VARCHAR (Variable-Length String):**
    *   Stores a variable-length string. It only uses the space required by the actual string, plus 1 or 2 bytes to record the length of the string.
    *   More efficient for storing strings of varying lengths, as it conserves disk space.
    *   **Practical Example:** Storing names (e.g., 'Alice', 'Christopher'). If you define `VARCHAR(50)`, 'Alice' will use 5 bytes + length byte, while 'Christopher' will use 11 bytes + length byte. Both would use up to 50 bytes if necessary.

**Differentiating Example:**
If you have a column for `gender` and you expect only 'M' or 'F', `CHAR(1)` is ideal because the length is always fixed. If you have a column for `comments` where text length can vary greatly, `VARCHAR(255)` or `TEXT` is more suitable to save space.

### 5. What is the purpose of NOT NULL and UNIQUE constraints?

Both `NOT NULL` and `UNIQUE` are column constraints used to enforce data integrity in a database table.

*   **NOT NULL Constraint:**
    *   **Purpose:** Ensures that a column cannot contain `NULL` values. This means that every row in the table must have a definite value for that column; it cannot be left empty or undefined.
    *   **Example:** An `emp_name` column typically has a `NOT NULL` constraint because every employee should have a name.

*   **UNIQUE Constraint:**
    *   **Purpose:** Ensures that all values in a column (or a set of columns) are distinct. No two rows can have the same value for that column.
    *   **Note:** A `UNIQUE` column *can* contain `NULL` values, and multiple rows can have `NULL` in a `UNIQUE` column (as `NULL` is not considered equal to `NULL` in SQL comparisons, though some database systems might treat this differently, MySQL allows multiple `NULL`s in a `UNIQUE` column by default).
    *   **Example:** An `email` column often has a `UNIQUE` constraint to ensure that each employee has a distinct email address.

Below are the SQL commands to perform the requested tasks. We will use Python functions to simulate MySQL output formatting for each command.

```python
# Helper function to simulate MySQL output formatting
def format_mysql_table(headers, data):
    if not headers or not data:
        return "Empty set (0.00 sec)"

    # Calculate maximum column widths
    col_widths = [len(header) for header in headers]
    for row in data:
        for i, item in enumerate(row):
            col_widths[i] = max(col_widths[i], len(str(item)))

    # Create separator line
    separator = "+" + "+".join([f"-{
            '-' * width
        }" for width in col_widths]) + "+"

    # Create header line
    header_line = "|" + "|".join([f" {header:<{col_widths[i]}} " for i, header in enumerate(headers)]) + "|"

    # Create data lines
    data_lines = []
    for row in data:
        data_lines.append(
            "|" + "|".join([f" {str(item):<{col_widths[i]}} " for i, item in enumerate(row)]) + "|"
        )

    return "\n".join([separator, header_line, separator] + data_lines + [separator])

def simulate_query_ok():
    return "Query OK, 0 rows affected (0.00 sec)"

def simulate_database_changed():
    return "Database changed"

def simulate_rows_affected(count):
    return f"Query OK, {count} rows affected (0.00 sec)\nRecords: {count}  Duplicates: 0  Warnings: 0"

print("MySQL Output Simulation Helper Functions Loaded.")
```
