## Part A: Theory Answers

1.  **What is the difference between DATE and DATETIME?**
    *   **DATE**: Stores only the date part (year, month, day) in `YYYY-MM-DD` format. It does not include time information.
    *   **DATETIME**: Stores both the date and time parts (`YYYY-MM-DD HH:MM:SS` format). It includes year, month, day, hour, minute, and second.

2.  **Why should email or phone number fields often be marked as UNIQUE?**
    *   Marking email or phone number fields as `UNIQUE` ensures that no two records in the table can have the same value for that field. This is crucial for:
        *   **Data Integrity**: Prevents duplicate entries for a unique identifier like an email address or phone number, ensuring each user or entity has a distinct contact point.
        *   **User Identification**: Allows these fields to act as alternative unique identifiers for logging in or retrieving accounts, without being the primary key.
        *   **Consistency**: Helps maintain a clean and reliable dataset by preventing accidental or erroneous duplication of contact information.

3.  **Can a table have multiple UNIQUE constraints? Explain.**
    *   Yes, a table can have multiple `UNIQUE` constraints. Each `UNIQUE` constraint can be applied to a single column or a combination of columns (a composite `UNIQUE` key). For example, a `students` table could have `email` as a `UNIQUE` constraint and `(first_name, last_name, date_of_birth)` as another composite `UNIQUE` constraint to prevent identical students from being entered. Each `UNIQUE` constraint guarantees uniqueness within its specified column(s) independently.

4.  **What happens if you try to insert a NULL value into a NOT NULL column?**
    *   If you try to insert a `NULL` value into a column that is defined with the `NOT NULL` constraint, the database will raise an error. The insertion operation will fail, and the row will not be added to the table. This constraint ensures that the specified column always contains a value.

5.  **Why is proper data type selection important while designing tables?**
    *   Proper data type selection is crucial for several reasons:
        *   **Data Integrity**: Ensures that the data stored is consistent with its intended use (e.g., preventing text in a numeric column).
        *   **Storage Efficiency**: Using the most appropriate (and often smallest) data type saves disk space, which can be significant in large databases.
        *   **Performance**: Queries and operations on columns with appropriate data types are generally faster. For example, comparing integers is quicker than comparing strings.
        *   **Data Validation**: Data types enforce basic validation rules, helping to prevent invalid data from being inserted.
        *   **Functionality**: Some functions and operations are only available or make sense for specific data types (e.g., date calculations on `DATE` columns).

````python
# Helper function to simulate MySQL output formatting
def format_mysql_table(headers, data):
    if not headers or not data:
        return "Empty set (0.00 sec)"

    # Calculate maximum column widths
    col_widths = [len(header) for header in headers]
    for row in data:
        for i, item in enumerate(row):
            col_widths[i] = max(col_widths[i], len(str(item)))

    # Create separator line
    separator = "+" + "+".join([f"-{
            '-' * width
        }" for width in col_widths]) + "+"

    # Create header line
    header_line = "|" + "|".join([f" {header:<{col_widths[i]}} " for i, header in enumerate(headers)]) + "|"

    # Create data lines
    data_lines = []
    for row in data:
        data_lines.append(
            "|" + "|".join([f" {str(item):<{col_widths[i]}} " for i, item in enumerate(row)]) + "|"
        )

    return "\n".join([separator, header_line, separator] + data_lines + [separator])

def simulate_query_ok():
    return "Query OK, 0 rows affected (0.00 sec)"

def simulate_database_changed():
    return "Database changed"

def simulate_rows_affected(count):
    return f"Query OK, {count} rows affected (0.00 sec)\nRecords: {count}  Duplicates: 0  Warnings: 0"

print("MySQL Output Simulation Helper Functions Loaded.")
````
