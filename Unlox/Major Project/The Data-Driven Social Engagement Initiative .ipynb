### Potential Performance and Engagement Metrics

Based on the `df.info()` output, here's a categorization of columns into potential performance and engagement metrics, along with their relevance:

#### Potential Performance Metrics:
These metrics directly measure business outcomes, sales efficiency, and profitability.

*   **Total_Sales**: Directly measures the total monetary value of sales for a transaction. This is a primary indicator of overall sales performance.
*   **Revenue**: Represents the net income after discounts, reflecting the actual earnings from sales. This is a key profitability metric.
*   **Units_Sold**: Indicates the quantity of products sold in a transaction. It's a measure of sales volume and product popularity.
*   **Unit_Price**: Shows the price of a single unit of a product. Can be used to analyze pricing strategies and their impact on sales.
*   **Discount_Percentage**: The percentage of discount applied to a product. Useful for understanding pricing elasticity and promotional effectiveness on revenue.
*   **Stock_On_Hand**: Represents the inventory level of products. This can indicate inventory management efficiency and potential for sales.
*   **Store_Rating**: An overall rating for the store, which can be an indicator of customer satisfaction and directly influence sales performance.

#### Potential Engagement Metrics:
These metrics provide insights into customer behavior, interaction, satisfaction, and the effectiveness of marketing or store-related factors.

*   **Customer_Type** (New/Returning): Distinguishes between new and repeat customers, which is crucial for understanding customer loyalty and retention—key aspects of customer engagement.
*   **Payment_Mode**: The method of payment used by customers. Analyzing payment modes can reveal customer preferences and convenience, potentially influencing engagement.
*   **Promotion_Applied**: Indicates whether a promotion was applied to a sale. This is a direct measure of promotional campaign effectiveness and customer response to marketing efforts.
*   **Product_Category** / **Product_Subcategory** / **Brand**: These provide insights into customer preferences and interests, indicating what products or brands engage customers most.
*   **Store_Location**: Can help understand regional customer engagement and preferences, indicating if certain locations draw more interest.
*   **Region**: Similar to Store_Location, it helps in understanding geographical customer engagement and tailoring strategies.
*   **Holiday_Flag**: Indicates if the transaction occurred on a holiday. This can show how holidays impact customer purchasing behavior and engagement with promotions.
